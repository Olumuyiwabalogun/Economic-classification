Project Title: Analyzing Economic Classification of Countries Based on GDP (PPP) and GDP Per Capita in 2022

Objective: This project aims to analyze the economic classification of countries by examining GDP (Purchasing Power Parity) and GDP per capita for the year 2022. Using a dataset sourced from the UN and IMF, the analysis involves data cleaning to ensure accuracy, followed by univariate and bivariate analysis techniques. Univariate analysis will describe individual economic indicators, while bivariate analysis will explore the relationship between GDP (PPP) and GDP per capita. Through this analysis, insights into the economic standings and disparities among different countries will be provided.


```python
# import the neccessary python libraries and open the target data set 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# laoding of the data set
df = pd.read_csv(r'C:\Users\Neos\Desktop\DATA.A\world economy.csv')
df
```

CLEANING OF DATASET

```python
df2 = df.dropna()
```

```python
df2.head()
```

```python
df2.tail()
```

```python
df2.drop_duplicates()
```

```python
df2.isnull().sum()
```

```python
def clean_gdp_column(column):
    #  column is treated as string before replacing
    column = column.astype(str)
    # to Remove dollar signs and commas
    cleaned_column = column.replace({r'\$': '', ',': ''}, regex=True)
    # to Convert to float
    cleaned_column = cleaned_column.astype(float)
    return cleaned_column

# Applying the function to clean the GDP columns
df2.loc[:, 'gdp_ppp_2022'] = clean_gdp_column(df2['gdp_ppp_2022'])
df2.loc[:, 'gdp_pc_2022'] = clean_gdp_column(df2['gdp_pc_2022'])
```

Univariate Analysis

Summary Statistics

```python
gdp_ppp = df2['gdp_ppp_2022']
gdp_ppp.describe()
```

```python
gdp_pc = df2['gdp_pc_2022']
gdp_ppp.describe()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df2['gdp_ppp_2022'], kde=True)
plt.title('GDP (PPP) 2022 Distribution')
```

```python
plt.figure(figsize=(8, 6))
plt.subplot(1, 2, 2)
sns.histplot(df2['gdp_pc_2022'], kde=True)
plt.title('GDP per Capita 2022 Distribution')
plt.show()
```

Bivariate Analysis

```python
print(df2[['gdp_ppp_2022', 'gdp_pc_2022']].describe())
```

```python
# Correlation
correlation = df2['gdp_ppp_2022'].corr(df2['gdp_pc_2022'])
print(f'Correlation between GDP PPP and GDP per capita: {correlation}')
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df2['gdp_ppp_2022'], kde=True, label='GDP PPP 2022')
sns.histplot(df2['gdp_pc_2022'], kde=True, label='GDP per Capita 2022', color='orange')
plt.legend()
plt.title('Distribution of GDP PPP 2022 and GDP per Capita 2022')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='gdp_ppp_2022', y='gdp_pc_2022', data=df2)
plt.title('GDP PPP 2022 vs GDP per Capita 2022')
plt.xlabel('GDP PPP 2022')
plt.ylabel('GDP per Capita 2022')
plt.show()
```

```python
# Joint plot with regression line
sns.jointplot(x='gdp_ppp_2022', y='gdp_pc_2022', data=df2, kind='reg')
plt.show()
```
